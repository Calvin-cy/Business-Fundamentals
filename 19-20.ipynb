```python
df19 = pd.read_csv('2019 most_streamed_tracks - spotify_global_2019_most_streamed_tracks_audio_features.csv')
```

```python
df19.head()
```

```python
df19['Title'] = df19['Track Name']
```

```python
del df19['Track Name']
```

```python
df19.info()
```

```python
total_streams19 = df19.groupby(['Title']).sum().sort_values('Streams',ascending= False).reset_index()
```

```python
total_streams19
```

```python
total_streams19 = total_streams19[['Title','Streams']]
```

```python
total_streams19.drop_duplicates(subset = ["Title"],inplace=True)
```

```python
total_streams19
```

```python
df2019 = total_streams19.merge(df2,'inner',on = 'Title')
```

```python
df2019.drop_duplicates(subset=['Title'],inplace=True)
```

```python
df2019.reset_index(inplace=True)
```

```python
df2019
```

```python
del df2019['index']
```

```python
df2019
```

```python
c_df2019 = df2019[['Title','artists','Streams','acousticness','duration_ms','energy','instrumentalness','key','liveness','loudness','mode','release_date','speechiness','tempo','valence']]
```

```python
df1920 = pd.concat([c_df2019,c_df2020])
```

```python
df1920
```

```python
df1920['release_date'] = pd.to_datetime(df1920['release_date'])
```

```python
df1920.info()
```

```python
df = df1920 
```

```python
df['Year'] = df['release_date'].apply(lambda x:x.year)
```

```python
df['Month'] = df['release_date'].apply(lambda x:x.month)
```

```python
df['WeekDay'] = df['release_date'].apply(lambda x:x.weekday())
```

```python
df['Week'] = df['release_date'].apply(lambda x:x.week)
```

```python
df.head()
```

```python
df
```

```python
df.columns
```

```python
df = df.groupby(['Title'],as_index= False).agg({'Streams':'sum','artists':'first','acousticness':'first',
                                           'duration_ms':'first','energy':'first','instrumentalness':'first',
                                           'key':'first','liveness':'first','loudness':'first','mode':'first',
                                           'release_date':'first','speechiness':'first','tempo':'first',
                                           'valence':'first','Year':'first','Month':'first','WeekDay':'first',
                                           'Week':'first'
                                          })
```
